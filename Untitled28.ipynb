## **One-Hot Encoding:**

Como os modelos de Machine Learning (ML) baseiam-se em cálculos matemáticos para realizar inferências, é necessário transformarmos variáveis categóricas em valores numéricos. Isso permite que os algoritmos "enxerguem" e processem categorias que, originalmente, estão em formato de texto.

O One-Hot Encoding consiste em identificar cada valor distinto dentro de uma coluna e transformá-lo em uma nova coluna exclusiva. Essas novas colunas são preenchidas com os valores binários 1 ou 0, dependendo da presença da informação original.

Por exemplo:

Imagine uma coluna chamada Nacionalidade com três valores distintos: Brasileira, Argentina e Chilena. Ao aplicarmos o One-Hot Encoding, o dataset é reestruturado:

DF original

Captura de ecrã de 2026-01-15 19-55-39.png



Df One-Hot Enconding

Captura de ecrã de 2026-01-15 19-56-38.png


Na prática, o algoritmo faz uma pergunta binária para cada linha: "Esta pessoa é brasileira?". Se sim, atribui 1; caso contrário, 0. Esse processo é repetido para todas as categorias existentes.

Dessa forma, convertemos dados categóricos em uma matriz numérica que preserva a informação original, mas agora em um formato que permite a aplicação de operações matemáticas e o treinamento de modelos de ML.


       

```python
# Criando variáveis dependentes e independentes
X = df.drop(['cancelou_cliente'], axis=1)
y = df['cancelou_cliente']
```

```python
# Importando sklearn para One-Hot
from sklearn.compose import make_column_transformer # make_column_transformer para realizar a transformação nas colunas
from sklearn.preprocessing import OneHotEncoder     # OneHotEncoder transformação a ser aplicada
```

```python
# Pegando colunas categóricas
colunas = ['genero',  'parceiro', 'dependentes',
       'servico_telefone', 'multiplas_linhas', 'provedor_internet',
       'seguranca_online', 'backup_online', 'protecao_dispositivo',
       'suporte_tecnico', 'streaming_tv', 'streaming_filmes', 'tipo_contrato',
       'fatura_digital', 'metodo_pagamento']
```

```python
# Transformador
one_hot = make_column_transformer(
    (OneHotEncoder(drop='first'), colunas),
    remainder='passthrough',
    sparse_threshold=0
)
```

make_column_transformer(...)
É uma função de conveniência que cria um objeto ColumnTransformer. Ela permite que você aplique transformações (como o One-Hot Encoding) apenas em colunas selecionadas, mantendo o restante do dataset intacto.

drop='first':
Elimina  o risco de multicolinearidade para qualquer modelo, Isso removerá a primeira categoria de cada grupo, independentemente de serem 2 ou 10 categorias.

remainder='passthrough'
Por padrão, o ColumnTransformer descarta qualquer coluna que não tenha sido listada na transformação.

Ao usar 'passthrough', você está dizendo: "As colunas que eu não mencionei (como 'valor_mensal' ou 'meses_na_empresa'), deixe-as passar sem nenhuma alteração".

sparse_threshold=0
Este argumento controla o formato de saída dos dados.

Matriz Esparsa (Sparse Matrix): É um formato de memória otimizado para quando você tem muitos zeros. É eficiente para o computador, mas impossível de visualizar como uma tabela (DataFrame).

sparse_threshold=0: Ao definir como zero, você força o Scikit-Learn a sempre retornar uma Matriz Densa (um array comum do NumPy).

Vantagem: Isso facilita muito a visualização dos dados transformados e a conversão de volta para um DataFrame do Pandas.

## **Treino e Teste (Train/Test)**

O conceito de Aprendizado de Máquina baseia-se no princípio de que o modelo identifica padrões em dados existentes para, a partir deles, realizar previsões. No entanto, antes de disponibilizarmos um modelo para produção, precisamos validá-lo e obter métricas sobre o seu desempenho real.

Por esse motivo, em vez de entregarmos a base de dados inteira para o algoritmo, dividimos o conjunto em pelo menos duas partes:

Conjunto de Treino: É a parte dos dados utilizada para que o modelo aprenda os padrões.

Conjunto de Teste: É uma parcela que o modelo não conhece durante o aprendizado. Ela serve apenas para avaliar o desempenho e apurar as métricas finais.

Analogia Prática: Pense no modelo como um aluno, no conjunto de treino como o material de estudo e no conjunto de teste como a prova final. Se o aluno tiver acesso às questões da prova enquanto estuda, ele pode simplesmente memorizá-las (o que chamamos de overfitting) em vez de aprender a lógica por trás do assunto.

Geralmente, o dataset é dividido reservando-se entre 15% a 30% para teste e o restante para treino. Embora existam divisões mais complexas (como a inclusão de um conjunto de validação), o conceito fundamental permanece o mesmo: garantir que o modelo seja testado com dados inéditos.

```python
# Importando método de divisão do grupo de dados
from sklearn.model_selection import train_test_split
```

Nós Iremos realizar a divisão do nossos dados em 3 grupos, treino, teste e validação. Dessa forma teremos uma

Ao final da divisão teremos a seguinte distribuição:

Treino ($X_{train}$): Onde o algoritmo busca padrões.

Validação ($X_{val}$): Onde testa se os padrões funcionam em dados "novos" enquanto ainda está desenvolvendo.

Teste ($X_{teste}$): O juiz final que dirá se o modelo está pronto para o mundo real.

```python
# Primeira divisão
X,X_teste,y,y_teste = train_test_split(X,y,test_size=0.15,stratify=y, random_state=5)
```

O parâmetro test_size separa 15% dos dados originais para o conjunto de Teste e o restante sobrescreve as variáveis $X$ e $y$ originais. Agora, elas contêm apenas os 85% restantes que sobraram do teste.


```python
# Segunda divisão
X_train,X_val,y_train,y_val = train_test_split(X,y,stratify=y, random_state=5)
```

Pega aqueles 85% que sobraram e os divide novamente.
Como não definiu test_size, o Scikit-Learn usa o padrão de 25% para o conjunto de Validação ($X_{val}$) e 75% para o de Treino ($X_{train}$).

Aqui a validação vai funcionar como um simulado da prova

Os Parâmetros-Chave:

stratify=y: Isso é fundamental. Ele garante que a proporção das classes (ex: quantos % cancelaram e quantos % ficaram) seja mantida exatamente igual em todos os novos conjuntos. Sem isso, você corre o risco de o treino ter muitos clientes que ficaram e a validação ter quase nenhum, enviesando o modelo.

random_state=5: É a "semente" da aleatoriedade. Garante que, se você rodar o código de novo, a divisão será exatamente a mesma. Isso permite que outros (ou o seu eu do futuro) consigam reproduzir seus resultados.

## **Modelagem**

Com nossos dados preparados e o Benchmark definido, podemos dar início à nossa modelagem. O primeiro algoritmo que utilizaremos será a Regressão Logística.

**O que é a Regressão Logística?**

A lógica deste modelo baseia-se no cálculo da probabilidade de um determinado evento acontecer (no nosso caso, o cancelamento do cliente) utilizando a função matemática sigmoide. Esta função mapeia qualquer valor real para um intervalo entre 0 e 1, permitindo análises probabilísticas.


Função Sigmoid :

$$f(x) = \frac{1}{1 + e^{-x}}$$

Onde:
$x$: Entrada de dados (as features do cliente).
$e$: Base do logaritmo natural (aproximadamente 2,718).

![Gráfico da Função Sigmoide](https://upload.wikimedia.org/wikipedia/commons/thumb/8/88/Logistic-curve.svg/320px-Logistic-curve.svg.png)

Tipos de Regressão Logística


Existem três variações principais deste modelo, dependendo da natureza do problema:

Binária: Utilizada para previsões entre duas possibilidades exclusivas, como diagnosticar se uma doença está presente ou ausente (Sim/Não). É este o tipo que utilizaremos em nosso projeto de Churn.

Multinomial: Realiza previsões quando há três ou mais categorias possíveis sem uma ordem inerente. Exemplo: prever qual meio de transporte um grupo de pessoas escolherá (ônibus, carro ou avião).

Ordinal: Aplicada quando há mais de duas categorias e a ordem entre elas é relevante. Exemplo: prever a classificação da qualidade de um atendimento em uma central (Ruim, Regular, Bom).

Para o nosso projeto iremos utilisar o tipo binário.

## **Árvores de Decisão**

O modelo de Árvore de Decisão funciona por meio de nós de decisão. O algoritmo divide os dados em nós, onde cada nó representa um subconjunto menor do grupo original. Esses subconjuntos são divididos sucessivamente até que o modelo apresente o resultado final. Por essa razão, podemos relacionar a lógica do modelo à estrutura de fluxogramas.

As Árvores de Decisão podem ser utilizadas em dois tipos de tarefas:

Classificação: Como no projeto em que estamos trabalhando, onde tentamos prever se um cliente irá cancelar ou não (Churn).

Regressão: Para previsões de valores contínuos, como a nota de um aluno em um teste final com base em seus hábitos de estudo.

A Diferença Fundamental
A principal diferença reside no tipo de saída: na classificação, o resultado é categórico (como "Sim" e "Não" ou "Aprovado" e "Reprovado"). Já na regressão, os resultados são valores numéricos contínuos (como peso, altura, custo, etc.).

```python
model_tree = DecisionTreeClassifier(max_depth=3,random_state=5)
model_tree.fit(X_train, y_train)
# verificando desempenho do modelo
print(f'Acurácia Treino: {model_tree.score(X_train,y_train)}')
print(f'Acurácia Validação: {model_tree.score(X_val,y_val)}')
```

Podemos observar que em nível de acurácia, a Regressão se apresentou ligeiramente melhor e que ambos os modelos generalisam bem para os dados de validação.

## **Random Forest**

O algoritmo Random Forest utiliza um conjunto de múltiplas Árvores de Decisão criadas de forma independente e aleatória. A decisão final do modelo não depende de apenas uma árvore, mas sim do consenso de todas elas.

Como funciona?
Imagine que, em vez de consultar apenas um especialista (uma única árvore), você convoque um conselho de especialistas. Cada árvore da "floresta" analisa os dados e dá o seu veredito.

Em problemas de classificação: O resultado final é definido pela votação da maioria.

Em problemas de regressão: O resultado final é a média das previsões de todas as árvores.

Principais Benefícios
Redução de Overfitting: Como o resultado é uma média de várias árvores, o modelo tende a ser muito mais estável e menos propenso a "decorar" o ruído dos dados de treino do que uma única árvore isolada.

Importância das Variáveis: O algoritmo facilita a identificação de quais variáveis (como tipo de contrato ou valor mensal) têm maior peso na decisão final.

Robustez: Lida bem com grandes conjuntos de dados e variáveis faltantes.

Aplicações
Devido à sua alta precisão, é amplamente aplicado em setores como:

Financeiro: Detecção de fraudes e análise de crédito.

Saúde: Diagnóstico de doenças e análise de histórico de pacientes.

E-commerce: Sistemas de recomendação e previsão de churn (como o nosso projeto).


```python
# Importando modelo

from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier(max_depth=3, n_estimators=100, random_state=5)
model_rf.fit(X_train, y_train)
# verificando desempenho do modelo
print(f'Treino: {model_rf.score(X_train, y_train)}')
print(f'Validação: {model_rf.score(X_val, y_val)}')
```

Isso nos ensina que modelos de conjunto (Ensembles) muitas vezes precisam de um pouco mais de 'espaço' para crescer e capturar a complexidade dos dados que modelos lineares como a Regressão Logística pegam de primeira.

## **Overfitting e Underfitting**

Antes de seguirmos,  é fundamental entendermos o conceito de Overfitting, mencionado anteriormente.

**Overfitting (Sobreajuste)**

O Overfitting ocorre quando o modelo se ajusta excessivamente aos dados de treinamento. Com isso, ao ser apresentado a dados desconhecidos, ele apresenta um desempenho muito abaixo do esperado, pois não consegue generalizar os padrões para novas informações.

Um indicador claro de overfitting é quando a acurácia do treinamento fica muito próxima ou igual a $1.0$ ($100\%$), enquanto o score para os dados de teste/validação é significativamente menor. Isso indica que o modelo "decorou" os dados de treino em vez de aprender as regras por trás deles.

Por esse motivo, realizamos o cálculo da acurácia tanto para o treino quanto para a validação simultaneamente.Underfitting (Subajuste)

Outra possibilidade que podemos encarar durante a modelagem é o Underfitting.

Nesse caso, o modelo é simples demais e não consegue capturar corretamente os padrões dos dados, resultando em pontuações de acurácia muito baixas em ambos os conjuntos (treino e teste). É como se o modelo não tivesse "estudado" o suficiente para entender o problema.

$y\_teste$: O "gabarito", contendo os dados reais conhecidos.

$y\_previsto$: O resultado da predição do modelo.

**Entendendo as Métricas de Validação**

Antes de escolhermos o melhor modelo, precisamos definir os critérios matemáticos. A base de qualquer métrica de classificação é a relação entre o que o modelo previu e o que realmente aconteceu:

TP (Verdadeiro Positivo): O modelo previu que o cliente cancelaria e ele realmente cancelou.

TN (Verdadeiro Negativo): O modelo previu que o cliente ficaria e ele realmente ficou.

FP (Falso Positivo): O modelo previu que o cliente cancelaria, mas ele permaneceu (Erro Tipo I).

FN (Falso Negativo): O modelo previu que o cliente ficaria, mas ele cancelou (Erro Tipo II).

Principais Métricas

Acurácia:

Mede a proporção global de acertos do modelo. É fácil de compreender, mas pode ser traiçoeira em conjuntos de dados desbalanceados.

$$\text{Acurácia} = \frac{TP + TN}{TP + TN + FP + FN}$$


Precisão: Proporção de previsões positivas que estão corretas. É útil quando o custo de um Falso Positivo é muito alto para a empresa.

$$\text{Precisão} = \frac{TP}{TP + FP}$$

Recall (Revocação): É a capacidade do modelo de "enxergar" os casos reais. Mede quantos, dentre todos os clientes que realmente cancelaram, o modelo foi capaz de capturar.

$$\text{Recall} = \frac{TP}{TP + FN}$$


F1-Score: Média harmônica entre Precisão e Recall. É uma excelente opção quando desejamos um equilíbrio entre Falsos Positivos e Falsos Negativos.

$$F1\text{-Score} = 2 \cdot \frac{\text{Precisão} \cdot \text{Recall}}{\text{Precisão} + \text{Recall}}$$

A escolha da mérica depende do problema de negócio e como as métrica pesam no resultado fina.


Para o nosso problema de negócio, o Recall é a métrica prioritária.
No contexto de retenção de clientes, o custo do erro é assimétrico:

Alto Recall: O modelo identifica quase todos os clientes em risco de saída. Isso permite que a empresa aja preventivamente na grande maioria dos casos.

O Custo do Falso Negativo ($FN$): Perder um cliente fiel para sempre por não ter previsto sua saída é muito mais caro do que oferecer um benefício para um cliente que talvez não cancelasse (Falso Positivo).

```python
# Cálculo de metricas

from sklearn.metrics import classification_report

print("========= RELATÓRIO DE MÉTRICAS =========")

print("\n1. REGRESSÃO LOGÍSTICA:")
print(classification_report(y_val, model_rl.predict(X_val), target_names=['Ficou', 'Cancelou']))

print("\n2. ÁRVORE DE DECISÃO:")
print(classification_report(y_val, model_tree.predict(X_val), target_names=['Ficou', 'Cancelou']))

print("\n3. RANDOM FOREST:")
print(classification_report(y_val, model_rf.predict(X_val), target_names=['Ficou', 'Cancelou']))
```

Após analisarmos os três modelos, a Regressão Logística se consolidou como a melhor escolha estratégica.

Enquanto a Árvore de Decisão e a Random Forest se mostraram modelos mais 'conservadores' — com alta precisão, mas baixo alcance (Recall) —, a Regressão Logística foi o modelo que melhor conseguiu 'caçar' os clientes em risco de cancelamento.

Em um cenário de Churn, identificar o maior número possível de clientes em risco (Recall) é mais valioso do que ser extremamente preciso em poucos casos. Com 55% de Recall, a Regressão Logística oferece o melhor ponto de partida para as campanhas de retenção da empresa.

Agora que já temos uma modelo definido vamos aplicar a prova final nele e extrair nossas conclusões finais

```python
# 1. Fazendo a previsão final com o campeão (Regressão Logística)
y_pred_final = model_rl.predict(X_teste)

# 2. Gerando o Relatório de Classificação Final
print("========= VEREDITO FINAL: DADOS DE TESTE (UNSEEN DATA) =========")
print(classification_report(y_teste, y_pred_final, target_names=['Ficou', 'Cancelou']))

# 3. Plotando a Matriz de Confusão Final
plt.figure(figsize=(8, 6))
cm_final = confusion_matrix(y_teste, y_pred_final)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_final, display_labels=['Ficou', 'Cancelou'])
disp.plot(cmap='Greens') # Usei Verde para simbolizar o "Selo de Aprovação"
plt.title('Matriz de Confusão Final: Regressão Logística (Dados de Teste)')
plt.show()
```

**Próximos Passos: Como podemos melhorar?**

Nenhum modelo é perfeito, e a Ciência de Dados é um processo iterativo. Se fôssemos continuar este projeto, aqui estão as frentes de melhoria que poderiam elevar ainda mais o nosso Recall e Precisão:

Ajuste de Threshold (Limiar de Decisão): Poderíamos alterar o limite de decisão da Regressão Logística para ser mais "agressivo". Ao baixar o limiar de 0.50 para 0.40, aumentaríamos o Recall (detectaríamos mais churns), embora isso pudesse gerar mais alarmes falsos.

Otimização de Hiperparâmetros: Utilizar ferramentas como GridSearchCV ou RandomizedSearchCV para encontrar os melhores parâmetros possíveis para a Random Forest, tentando superar o desempenho da Regressão Logística.

Teste de Modelos de Gradient Boosting: Implementar algoritmos como XGBoost ou LightGBM, que costumam ter performances superiores em dados tabulares complexos.
